```python

# Install necessary packages
!pip install datasets
from datasets import load_dataset
dataset = load_dataset('toughdata/quora-question-answer-dataset', split='train')

import pandas as pd
from huggingface_hub import login
login()
from datasets import load_dataset
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

# Download NLTK resources
nltk.download('stopwords')
nltk.download('punkt')

dataset = load_dataset('toughdata/quora-question-answer-dataset', split='train')

# Convert to DataFrame for easier manipulation
df = pd.DataFrame(dataset)

# Print the column names to find the correct ones
print(df.columns)

# Data Cleaning and Preprocessing
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()

def preprocess_text(text):
    # Tokenization
    tokens = word_tokenize(text.lower())
    # Remove stop words and apply stemming
    filtered_tokens = [stemmer.stem(w) for w in tokens if w.isalnum() and w not in stop_words]
    return ' '.join(filtered_tokens)

# Assuming the columns are 'question' and 'answer' based on inspection
df['processed_questions'] = df['question'].apply(preprocess_text)
df['processed_answers'] = df['answer'].apply(preprocess_text)

# Display the processed data
print(df.head())
!pip install rouge-score rouge datasets transformers evaluate pandas nltk

import torch
from transformers import BertForQuestionAnswering, BertTokenizer, T5ForConditionalGeneration, T5Tokenizer, GPT2LMHeadModel, GPT2Tokenizer
import evaluate
from sklearn.metrics import f1_score

# Initialize models and tokenizers
bert_model = BertForQuestionAnswering.from_pretrained('bert-large-uncased-whole-word-masking-finetuned-squad')
bert_tokenizer = BertTokenizer.from_pretrained('bert-large-uncased-whole-word-masking-finetuned-squad')

t5_model = T5ForConditionalGeneration.from_pretrained('t5-small')
t5_tokenizer = T5Tokenizer.from_pretrained('t5-small')

gpt_model = GPT2LMHeadModel.from_pretrained('gpt2')
gpt_tokenizer = GPT2Tokenizer.from_pretrained('gpt2')

# Define evaluation function for BERT model
def evaluate_model(model, tokenizer, question, context):
    inputs = tokenizer.encode_plus(question, context, return_tensors='pt')
    outputs = model(input_ids=inputs['input_ids'], attention_mask=inputs['attention_mask'])
    start_scores = outputs.start_logits
    end_scores = outputs.end_logits
    all_tokens = tokenizer.convert_ids_to_tokens(inputs['input_ids'][0].tolist())
    answer = ' '.join(all_tokens[torch.argmax(start_scores) : torch.argmax(end_scores) + 1])
    return answer.replace(' ##', '')

# Evaluate BERT model
sample_question = df['processed_questions'].iloc[0]
sample_answer = df['processed_answers'].iloc[0]
bert_answer = evaluate_model(bert_model, bert_tokenizer, sample_question, sample_answer)
print(f"BERT Answer: {bert_answer}")

# Define evaluation function for T5 model
def evaluate_t5_model(model, tokenizer, question, context):
    input_text = f"question: {question} context: {context}"
    input_ids = tokenizer(input_text, return_tensors='pt').input_ids
    outputs = model.generate(input_ids)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Evaluate T5 model
t5_answer = evaluate_t5_model(t5_model, t5_tokenizer, sample_question, sample_answer)
print(f"T5 Answer: {t5_answer}")

# Define evaluation function for GPT-2 model
def evaluate_gpt2_model(model, tokenizer, question, context):
    input_text = f"Question: {question}\nContext: {context}\nAnswer:"
    input_ids = tokenizer.encode(input_text, return_tensors='pt')
    outputs = model.generate(input_ids, max_length=50, num_return_sequences=1)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Evaluate GPT-2 model
gpt_answer = evaluate_gpt2_model(gpt_model, gpt_tokenizer, sample_question, sample_answer)
print(f"GPT-2 Answer: {gpt_answer}")

# Initialize evaluation metrics
!pip install --upgrade evaluate
import evaluate
rouge = evaluate.load('rouge')
bleu = evaluate.load('bleu')
from datasets import load_metric
f1 = load_metric("f1")

# Compute and print metrics
def compute_metrics(preds, labels):
    rouge = evaluate.load('rouge')
    bleu = evaluate.load('bleu')
    rouge_scores = rouge.compute(predictions=preds, references=labels)
    bleu_scores = bleu.compute(predictions=preds, references=labels)
    f1_scores = [f1_score([label], [pred], average='weighted') for label, pred in zip(labels, preds)]
    return rouge_scores, bleu_scores, f1_scores

# Example usage
predictions = [bert_answer, t5_answer, gpt_answer]
references = [sample_answer, sample_answer, sample_answer]
rouge_scores, bleu_scores, f1_scores = compute_metrics(predictions, references)

print(f"ROUGE Scores: {rouge_scores}")
print(f"BLEU Scores: {bleu_scores}")
print(f"F1 Scores: {f1_scores}")

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go

plt.figure(figsize=(10, 6))
sns.histplot(df['processed_questions'].apply(lambda x: len(x.split())), bins=50, kde=True)
plt.title('Distribution of Question Lengths')
plt.xlabel('Number of Words')
plt.ylabel('Frequency')
plt.show()

# Feature Importance
plt.figure(figsize=(10, 6))
sns.histplot(df['processed_answers'].apply(lambda x: len(x.split())), bins=50, kde=True, color='orange')
plt.title('Distribution of Answer Lengths')
plt.xlabel('Number of Words')
plt.ylabel('Frequency')
plt.show()

# Model Performance Visualization
performance_data = {
    'Model': ['BERT', 'T5', 'GPT-2'],
    'ROUGE-1': [rouge_scores['rouge1'], rouge_scores['rouge1'], rouge_scores['rouge1']],
    'BLEU': [bleu_scores['bleu'], bleu_scores['bleu'], bleu_scores['bleu']],
    'F1': [f1_scores[0], f1_scores[1], f1_scores[2]]
}

performance_df = pd.DataFrame(performance_data)

fig = px.bar(performance_df, x='Model', y=['ROUGE-1', 'BLEU', 'F1'], barmode='group', title="Model Performance Metrics")
fig.show()
```
